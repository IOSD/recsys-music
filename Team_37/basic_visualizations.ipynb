```python
import numpy as np 
import seaborn as sns
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import matplotlib.pyplot as plt
```

```python
ratings = pd.read_csv('../input/ratings.csv')
ratings.head()
```

```python
users = ratings.userId.unique()
movies = ratings.movieId.unique()
n_user = len(users)
n_movie =  len(movies)
```

```python
fig = sns.set(rc={'figure.figsize':(13,5)})
ax = sns.kdeplot(ratings["userId"].value_counts().sort_values(ascending=False) , shade=True, color="aqua")
plt.xlabel("No. of movies rated")
plt.title("Rated movies Distribution")
plt.grid(True)
plt.show(fig)
```

```python
fig = sns.set(rc={'figure.figsize':(23,10)})
plt.subplot(221)
fig1 = sns.violinplot(x=ratings["rating"])
plt.xlabel("Ratings")
plt.title("Ratings Distribution (Violin)")
plt.grid(True)
plt.show(fig1)
plt.subplot(222)
fig2 = sns.kdeplot(ratings.rating , shade=True, color="r")
plt.xlabel("Ratings")
plt.title("Ratings Distribution (Kde)")
plt.grid(True)
plt.show(fig2)
```

```python
sns.set(rc={'figure.figsize':(13,5)})
ax = sns.kdeplot(ratings["movieId"].value_counts().sort_values(ascending=False) , shade=True, color="coral")
plt.xlabel("No. of times")
plt.title("Distribution of number of times a movie is rated")
plt.grid(True)
plt.show(ax)
```

```python
sns.set(rc={'figure.figsize':(13,5)})
ax = sns.countplot(x="rating", data=ratings , palette = "Set2")
plt.xlabel("Ratings")
plt.ylabel("Count of Occurences")
plt.title("Ratings count")
plt.grid(True)
plt.show(ax)
```
